```python
from psfmodels.v2 import _aguet, _li2017
from psfmodels import _cuvec
import _psfmodels
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import PowerNorm
import contextlib
import time

@contextlib.contextmanager
def timer(txt: str = ''):
    start = time.perf_counter_ns()
    yield
    end = time.perf_counter_ns()
    print(txt, f"elapsed time: {(end - start)/1000:.2f} us")
```

```python
zvec = np.linspace(-2, 2, 161)
nx = 151
dxy = 0.01
r_vec = np.arange(nx) * dxy  - (nx - 1) * dxy / 2

wvl = 0.6
pz = 0
NA = 1.1

params = dict(
    ng0 = 1.515,  # coverslip RI design value
    ng = 1.515,  # coverslip RI experimental value
    ni0 = 1.515,  # immersion medium RI design value
    ni = 1.515,  # immersion medium RI experimental value
    ns = 1.47,    # specimen refractive index (RI)
    ti0 = 150.0,  # microns, working distance (immersion medium thickness) design value
    tg = 170.0,  # microns, coverslip thickness experimental value
    tg0 = 170.0,  # microns, coverslip thickness design value
)


with timer('cpp code'):
    c_psf = _psfmodels.vectorial_psf(
        zv = zvec.copy(),
        nx = nx,
        dxy = dxy,
        pz = pz,
        wvl = wvl,
        NA = NA,
        **params,
        sf = 3,
        mode = 1
    )
    c_psf = c_psf[:, c_psf.shape[1]//2]
    print(c_psf.shape)
with timer('new aguet'):
    new_aguet = _aguet.radial_psf_aguet2009(z=zvec, r=r_vec, na=NA, **params)
    print(new_aguet.shape)
with timer('li'):
    li_psf = _li2017.radial_psf_li2017(z=zvec, r=r_vec, na=NA, **params)
    print(li_psf.shape)
with timer('_cuvec'):
    _params = {
        'na': NA,
        'coverslip_ri': params['ng'],
        'coverslip_ri_spec': params['ng0'],
        'immersion_medium_ri': params['ni'],
        'immersion_medium_ri_spec': params['ni0'],
        'specimen_ri': params['ns'],
        'working_distance': params['ti0'],
        'coverslip_thickness': params['tg'],
        'coverslip_thickness_spec': params['tg0'],
    }
#     cuv_psf = _cuvec.vectorial_rz(zv=zvec, nx=nx, dxy=dxy, wvl=wvl, params=_params)
    cuv_psf = _cuvec.vectorial_psf(zvec, nx=nx, dxy=dxy, params=_params)
    cuv_psf = cuv_psf[:, cuv_psf.shape[1]//2]
    print(cuv_psf.shape)
```

```python
fig, axs = plt.subplots(1, 4, figsize=(12,9))

axs[0].imshow(c_psf, norm=PowerNorm(0.5))
axs[1].imshow(new_aguet, norm=PowerNorm(0.5))
axs[2].imshow(li_psf, norm=PowerNorm(0.5))
axs[3].imshow(cuv_psf, norm=PowerNorm(0.5))
```

```python
import numpy.testing as npt

npt.assert_allclose(new_aguet/new_aguet.max(), new_aguet/new_aguet.max(), rtol=1e-8)
```
